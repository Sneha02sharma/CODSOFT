**<font size="5">Feature Scaling</font>**

```python
#Importing the necessary libraries for preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x=CreditCard_data.drop(["Class"],axis=1)
```

```python
y=CreditCard_data["Class"]
```

```python
#target/species data is stored in y
y
```

```python
#Shows starting 5 values of x
x.head()
```

```python
#Splitting the dataset in feature class and target variable for training and testing
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#X has data Amount and Class
X_train.shape
```

```python
#Checking the shape for Y for training set
y_train.shape
```

```python
#Checking the shape for Y in training set
y_test.shape
```

```python
#We'll do standard scaling for the given dataset
ss= StandardScaler()
ss.fit_transform(X_train)
ss.transform(X_test)
```

**<font size="5"> Random Forest Model</font>**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,classification_report,confusion_matrix
```

```python
#initialise  and train the random forest model
rf=RandomForestClassifier(class_weight='balanced', random_state=42)
#fit the training and testing data in Random forest model
rf.fit(X_train,y_train)
```

```python
#Now let's find the predicted value  for y
y_pred=rf.predict(X_test)
```

```python
#lets make a scatter plot to visualize our model outputs
plt.scatter(y_test,y_pred,alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("predicted")
plt.show()
```

```python
Here we can see two values are correctly predicted whereas two are incorrectly predicted (0,0), (1,1) are correctly predicted 
```

```python
#Accuracy with predicted and actual value
("ACCURACY=",accuracy_score(y_test,y_pred))
```

So,the accuracy is 0.99

```python
#precision score
precision_score(y_test,y_pred,average='weighted')
```

So,the precision is 0.99

```python
#recall score
recall_score(y_test,y_pred,average='weighted')
```

So,the recall score is 0.99

```python
#Confusion_matrix 
confusion_matrix(y_test,y_pred)
```

In the first case, 84966 values are correctly predicted in class 0 and 8 are incorrectly predicted,
In the second case , 106 values are correctly predicted in class 1 and 38 are incorrectly predicted,

```python
#Classification_report for the dataset
classification_report(y_test,y_pred)
```
